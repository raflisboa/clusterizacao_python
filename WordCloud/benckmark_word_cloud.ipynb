# Wordcloud (CSV) Model

Generating a wordcloud from csv source.

```python
# importar os pacotes necessários
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# importar o arquivo csv local em uma variável ("df")
#df = pd.read_csv("csv/dados_benchmark.csv")
#df = pd.read_csv("csv/ementas_benchmark.csv")
#df = pd.read_csv("csv/polos_tecnologicos.csv")
#df = pd.read_csv("csv/cases_pesquisa.csv")
#df = pd.read_csv("csv/outros_estados.csv")
#df = pd.read_csv('csv/levantamento_requisitos.csv', delimiter=',')
df = pd.read_csv('csv/soma_maioria.csv', delimiter=',')
#df = pd.read_csv('csv/carga_horaria.csv', delimiter=',')
```

```python
# Selecionar colunas como base da nuvem
summary = df.dropna(subset=['carga_horaria'], axis=0)['cursos']
```

```python
# array de palavras indesejadas a serem removidas
myStopwords = ["da", "meu", "em", "você", "de", "ao", "os", "\n", "por", "e", "da", "o", "é", "que","para", "cursos", "uma", "mais", "serão", "um", "além", "como", "pelo","das", "dos","aos", "as","na", "no","nas", "nos", "assim", "end", "oferecido", "á", "à", "através", "essa", "esse", "essas", "esses", "entre","tendo", "DF", "Certifique", "seu", "conhecimento", "considerações", "finais", "parte", "sobre", "none", "pagina","página", "objetivo", "tipo", "componente","aula","método","curso","está","também","oferece","pela","ampliando","seu","sua","apoio","vida","exemplo","conheça","seus","usando","forma","usar","mínima","anos","idade","minima","acesso", "alguns", "algum", "final","se", "estado", "região","empresas", "cidade", "grande", "foi", "maior","mil","onde","conta","tem","apelido","nesta","mão","segunda","localizada","frente","já","quatro","Pedro","meio","ou","será","tais","muito","ainda","Este","Neste", "uso", "escola", "escolas", "computador", "percepção", "alunos","realizada", "atividade", "sala", "número", "números","foram","compreender","vale","são", "trilhas","sistema", "trabalhando","criando", "objeto", "conteúdo", "aprenda", "usuário","comando","conteúdos", "IBM","SEDE","ampliar","proporcionar","gerando","lacunas","fechar","visando","requerido","sinergia","baixa","amplia","pessoas","teste","configuração","prática","Aperfeiçoamento"] + list(STOPWORDS)
```

```python
# validação da funcão anterior (stopwords)
myStopwords[10:35]
```

```python
# aplicar um colormap com base numa imagem prévia
mask = np.array(Image.open('img/Logo_FACTI.png'))
mask_colors = ImageColorGenerator(mask)
```

```python
# Nuvem de palavras com base no csv
# concatenar as palavras
all_summary = " ".join(s for s in summary)
 
# gerar a wordcloud a partir do array myStopwords
wordcloud = WordCloud(stopwords=myStopwords,
                      background_color="black", max_words=20,
                      width=3200, height=1600,colormap="Greens").generate(all_summary)
 
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(15,15))
ax.imshow(wordcloud, interpolation='sinc')
ax.set_axis_off()

# exportar a nuvem resultante
plt.imshow(wordcloud);
wordcloud.to_file("img/export/benckmark_wordcloud.png")
```

```python
# Nuvem de palavras com base no csv com template
# endereço local do template
image_mask = np.array(Image.open("img/new_facti.png"))
   
# gerar uma wordcloud
wordcloud = WordCloud(stopwords=myStopwords,
                      background_color="black",
                      colormap="Greens",
                      width=1600, height=1600, 
                      max_words=15,
                      mask=image_mask, max_font_size=200,
                      contour_color='#478e98',
                      contour_width=25,
                      min_font_size=5).generate(all_summary)
 
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(wordcloud, interpolation='sincx.set_axis_off()
 
plt.imshow(wordcloud)
wordcloud.to_file("img/export/factiBG_wordcloud.png")
```
