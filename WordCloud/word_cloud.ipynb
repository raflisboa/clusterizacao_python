```python
# importar os pacotes necessários
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# importar o arquivo csv da rede em um df
# df = pd.read_csv("http://dl.dropboxusercontent.com/s/fdzigk974zbx2de/listings.csv?dl=0")

# importar o arquivo csv local em um df
#df = pd.read_csv("csv/another.csv")
df = pd.read_csv("csv/ementas_benchmark.csv")

# eliminar as colunas com valores ausentes
#summary = df.dropna(subset=['summary'], axis=0)['summary']
#summary = df.dropna(subset=['firstname'], axis=0)['firstname']
summary = df.dropna(subset=['ementa'], axis=0)['ementa']
```

```python
# exemplos de retorno de dados
display(summary.iloc[0])
display(summary.iloc[1])
```

```python
# concatenar as palavras
all_summary = " ".join(s for s in summary)
 
# lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(["da", "meu", "em", "você", "de", "ao", "os"])
 
# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(all_summary)
 
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud);
#wordcloud.to_file("airbnb_summary_wordcloud.png")
wordcloud.to_file("test_wordcloud.png")
```

```python
# endereço LOCAL da SUA imagem
image_mask = np.array(Image.open("img/new_facti.png"))
   
# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="white",
                      width=1000, height=1000, max_words=2000,
                      mask=image_mask, max_font_size=200,
                      min_font_size=5).generate(all_summary)
 
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud)
#wordcloud.to_file("airbnb_summary_wordcloud.png")
wordcloud.to_file("test_summary_wordcloud.png")
```
